```python
import pandas as pd
```

```python
df=pd.read_csv('Ecommerce_data.csv')
```

```python
df
```

```python
df.label_num=df.label.map({
    'Household':0,
    'Electronics':1,
    'Clothing & Accessories':2,
    'Books':3,
})
```

```python
import spacy
import pandas as pd

# word vectors you need to install large or medium english model. We will install the large one!
# make sure you have run "python -m spacy download en_core_web_lg" to install large english model
nlp = spacy.load("en_core_web_lg")


def remove_stop_words_spacy(text):
    doc = nlp(text)
    filtered_text = ' '.join(token.text for token in doc if not token.is_stop)
    return filtered_text

# Apply the remove_stop_words_spacy function to the "Text" column
df['Text'] = df['Text'].apply(remove_stop_words_spacy)

df['Text'][0]
```

```python
#This will take some time(nearly 15 minutes)
df['vector'] = df['Text'].apply(lambda x: nlp(x).vector)  
```

```python
df['Text'][0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.vector.values,
    df.label_num,
    test_size=0.2
)
```
